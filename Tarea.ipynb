#README
Programa de figuras geometricas usando herencia y encapsulamiento.
calculo de area y perimetro de rectangulo, cuadrado y circunferencia

```python
class FiguraGeometrica:
    def __init__(self, alto, ancho):
        self._alto = alto
        self._ancho = ancho

    # Encapsulamiento con property
    @property
    def alto(self):
        return self._alto

    @alto.setter
    def alto(self, valor):
        if valor > 0:
            self._alto = valor

    @property
    def ancho(self):
        return self._ancho

    @ancho.setter
    def ancho(self, valor):
        if valor > 0:
            self._ancho = valor

    def __str__(self):
        return f"Alto: {self._alto}, Ancho: {self._ancho}"


class Color:
    def __init__(self, color):
        self._color = color

    @property
    def color(self):
        return self._color

    @color.setter
    def color(self, valor):
        self._color = valor

    def __str__(self):
        return f"Color: {self._color}"


# -------------------------------
# HERENCIA + SOBRESCRITURA
# -------------------------------

class Rectangulo(FiguraGeometrica, Color):
    def __init__(self, alto, ancho, color):
        FiguraGeometrica.__init__(self, alto, ancho)
        Color.__init__(self, color)

    def area(self):
        return self.alto * self.ancho

    def perimetro(self):
        return 2 * (self.alto + self.ancho)

    def __str__(self):
        return (
            f"Rectángulo -> {FiguraGeometrica.__str__(self)}, "
            f"{Color.__str__(self)}, Área: {self.area()}, "
            f"Perímetro: {self.perimetro()}"
        )


class Cuadrado(FiguraGeometrica, Color):
    def __init__(self, lado, color):
        super().__init__(lado, lado)
        Color.__init__(self, color)

    def area(self):
        return self.alto ** 2

    def perimetro(self):
        return 4 * self.alto

    def __str__(self):
        return (
            f"Cuadrado -> Lado: {self.alto}, {Color.__str__(self)}, "
            f"Área: {self.area()}, Perímetro: {self.perimetro()}"
        )


class Circunferencia(FiguraGeometrica, Color):
    def __init__(self, radio, color):
        super().__init__(radio, radio)
        Color.__init__(self, color)

    def area(self):
        import math
        return math.pi * (self.alto ** 2)

    def perimetro(self):
        import math
        return 2 * math.pi * self.alto

    def __str__(self):
        return (
            f"Circunferencia -> Radio: {self.alto}, "
            f"{Color.__str__(self)}, Área: {self.area():.2f}, "
            f"Perímetro: {self.perimetro():.2f}"
        )

if __name__ == "__main__":
    r = Rectangulo(12, 13, "Rojo")
    c = Cuadrado(10, "Azul")
    ci = Circunferencia(4, "Verde")

    print(r)
    print(c)
    print(ci)
```
